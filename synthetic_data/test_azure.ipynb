```python
import os
from openai import OpenAI

# Note this is only compatible with the completions API
# The Azure SDK is only compatible with the Responses API (which has been excluded here)
client = OpenAI(
    api_key=os.environ["AZURE_PROJECT_API_KEY"],
    base_url=os.environ["AZURE_OPENAI_ENDPOINT"] # https://YOUR-RESOURCE-NAME.openai.azure.com/openai/v1/
)

response = client.chat.completions.create(
    model="gpt-5-mini",
    messages=[
        {
            "role": "user",
            "content": "What's the capital of France?"
        }
    ]
)

print(response.choices[0].message.content)
```
